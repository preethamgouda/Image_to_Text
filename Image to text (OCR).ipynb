## Load the image

```python
img = PIL.Image.open('1.png')
plt.imshow(img)
```

## Convert Image to Text

```python
import pytesseract

pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract/tesseract'

TESSDATA_PREFIX = 'C:/Program Files/Tesseract'

```

```python
text_data = pytesseract.image_to_string(img.convert('RGB'), lang='eng')
```

```python
print(text_data)
```

## Extract Specific Fields

```python
import re

m = re.search("Name: (\w+)", text_data)
name = m[1]
print(name)
```

```python
m = re.search("Start Date: (\S+)", text_data)
start_date = m[1]
start_date
```

```python
m = re.search("Geo-Coordinates: (\S+)", text_data)
coordinates = m[1]
coordinates
```
